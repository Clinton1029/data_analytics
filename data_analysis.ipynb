```python
import pandas as pd   # pandas is used for data cleaning, manipulation, loading files, and DataFrame operations.
```

```python
# Load the Iris dataset into a pandas DataFrame

df = pd.read_csv("/content/Iris.csv")   # Loading CSV file
```

```python
# View first rows of the DataFrame
print("Head of the DataFrame:")
display(df.head())

# Select a single column
print("\nSelecting 'SepalLengthCm':")
display(df['SepalLengthCm'])

# Select multiple columns
print("\nSelecting 'SepalLengthCm' and 'PetalLengthCm':")
display(df[['SepalLengthCm', 'PetalLengthCm']])

# Select the 5th row (index 4)
print("\nSelecting 5th row:")
display(df.iloc[4])

# Filter rows where SepalLengthCm > 5
print("\nFiltering rows where SepalLengthCm > 5:")
display(df[df['SepalLengthCm'] > 5])
```

```python
# Removing a column (example: Id)
df_removed = df.drop("Id", axis=1)

print("DataFrame after removing 'Id' column:")
display(df_removed)
```

```python
# Renaming columns
df_renamed = df.rename(columns={
    "SepalLengthCm": "Sepal_Length",
    "SepalWidthCm": "Sepal_Width"
})

print("DataFrame after renaming SepalLengthCm and SepalWidthCm:")
display(df_renamed)
```

```python
# Using df.info()
print("Using df.info():")
df.info()

print("\nNumber of rows (len(df)):", len(df))

print("Number of columns (len(df.columns)):", len(df.columns))

print("Number of axes (len(df.axes)):", len(df.axes))  # rows & columns axis

print("Total size (df.size):", df.size)  # rows × columns

print("Shape of DataFrame (df.shape):", df.shape)  # (rows, columns)
```

```python
# First 5 rows
print("First 5 rows:")
display(df.head())

# Last 5 rows
print("\nLast 5 rows:")
display(df.tail())

# Random sample of 5 rows
print("\nRandom sample of 5 rows:")
display(df.sample(5))
```
